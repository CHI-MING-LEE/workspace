## 選股

```python
# 股價在均線之上
sampledates = {'gte':'2019-02-01','lte':'2019-05-15'}
data_raw = tejapi.get('TWN/APRCD', mdate=sampledates)
```

```python
close_df = data_raw.pivot("mdate", "coid")['close_d']
condition1 = close_df.iloc[-1] > close_df.iloc[-60:].mean()
condition1[condition1]
```
