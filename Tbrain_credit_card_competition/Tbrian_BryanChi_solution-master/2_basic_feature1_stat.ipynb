```python
## this notebook
#1. add day_num (ith day i=1...92)
#2. create day_cnt df, first column is fild_id feature is 0 , 1 , ... 92 indicate how many data corresponding to that file in that day_num
#3. create cust_cnt, prod_cnt, count how many customer_id, product_id, corresponding to that file
#4. create week_cnt, hour_cnt, min_cnt, sec_cnt count how many data corresponding to that time
#5. create prod_id_cnt count how many data corresponding to that product_id (divide by prod_sum)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gc

%matplotlib inline
```

# day_cnt

```python
#let's create daily dictionary
daily = list(np.arange(31)+1)
month = ['03','04','05']
dailydict = dict()
j=300
for i in daily:
    j=j+1
    dailydict[j]=i
j=400
for i in daily[:-1]:
    j=j+1
    dailydict[j]=i+31
j=500
for i in daily:
    j=j+1
    dailydict[j]=i+61
```

```python
march = pd.read_hdf('march.h5', 'df')
march['day_num']=march['tm_mon']*100 + march['tm_day']
march['day_num'] = march['day_num'].map(dailydict)
march.to_hdf('march.h5','df')
```

```python
april = pd.read_hdf('april.h5', 'df')
april['day_num']=april['tm_mon']*100 + april['tm_day']
april['day_num'] = april['day_num'].map(dailydict)
april.to_hdf('april.h5','df')
```

```python
may = pd.read_hdf('may.h5', 'df')
may['day_num']=may['tm_mon']*100 + may['tm_day']
may['day_num'] = may['day_num'].map(dailydict)
may.to_hdf('may.h5','df')
```

```python
march = pd.read_hdf('march.h5', 'df')
march_cnt = march[['file_id','day_num','timestamp']].groupby(['file_id','day_num']).count().reset_index()
march_cnt.columns = ['file_id','day_num','frequency']
march_cnt = march_cnt.pivot_table(values='frequency',index=['file_id'],columns='day_num',aggfunc='sum').reset_index()
del march
gc.collect();
```

```python
april = pd.read_hdf('april.h5', 'df')
april_cnt = april[['file_id','day_num','timestamp']].groupby(['file_id','day_num']).count().reset_index()
april_cnt.columns = ['file_id','day_num','frequency']
april_cnt = april_cnt.pivot_table(values='frequency',index=['file_id'],columns='day_num',aggfunc='sum').reset_index()

del april
gc.collect();
```

```python
may = pd.read_hdf('may.h5', 'df')
may_cnt = may[['file_id','day_num','timestamp']].groupby(['file_id','day_num']).count().reset_index()
may_cnt.columns = ['file_id','day_num','frequency']
may_cnt = may_cnt.pivot_table(values='frequency',index=['file_id'],columns='day_num',aggfunc='sum').reset_index()

del may
gc.collect();
```

```python
data = pd.merge(march_cnt,april_cnt,how='outer',on='file_id')
data = pd.merge(data,may_cnt,how='outer',on='file_id')
```

# week_cnt

```python
march = pd.read_hdf('march.h5', 'df')
march_cnt = march[['file_id','tm_wday','timestamp']].groupby(['file_id','tm_wday']).count().reset_index()
march_cnt.columns = ['file_id','tm_wday','frequency']
march_cnt = march_cnt.pivot_table(values='frequency',index=['file_id'],columns='tm_wday',aggfunc='sum').reset_index()
del march
gc.collect();

april = pd.read_hdf('april.h5', 'df')
april_cnt = april[['file_id','tm_wday','timestamp']].groupby(['file_id','tm_wday']).count().reset_index()
april_cnt.columns = ['file_id','tm_wday','frequency']
april_cnt = april_cnt.pivot_table(values='frequency',index=['file_id'],columns='tm_wday',aggfunc='sum').reset_index()

del april
gc.collect();

may = pd.read_hdf('may.h5', 'df')
may_cnt = may[['file_id','tm_wday','timestamp']].groupby(['file_id','tm_wday']).count().reset_index()
may_cnt.columns = ['file_id','tm_wday','frequency']
may_cnt = may_cnt.pivot_table(values='frequency',index=['file_id'],columns='tm_wday',aggfunc='sum').reset_index()

del may
gc.collect();
```

```python
data = pd.merge(march_cnt,april_cnt,how='outer',on='file_id')
data = pd.merge(data,may_cnt,how='outer',on='file_id')
```

```python
data = data.fillna(0)
data.columns = [str(i) for i in data.columns]
week_cnt_df = pd.DataFrame(data.file_id.unique())
week_cnt_df.columns = ['file_id']
for i in range(7):
    week_cnt_df[i]=0
    for feat in list(data.columns):
        if str(i) in feat:
            week_cnt_df[i]+=data[feat]
```

```python
week_cnt_df.to_csv('week_cnt.csv',index=False)
```

# hour_cnt

```python
march = pd.read_hdf('march.h5', 'df')
march_cnt = march[['file_id','tm_hour','timestamp']].groupby(['file_id','tm_hour']).count().reset_index()
march_cnt.columns = ['file_id','tm_hour','frequency']
march_cnt = march_cnt.pivot_table(values='frequency',index=['file_id'],columns='tm_hour',aggfunc='sum').reset_index()
del march
gc.collect();

april = pd.read_hdf('april.h5', 'df')
april_cnt = april[['file_id','tm_hour','timestamp']].groupby(['file_id','tm_hour']).count().reset_index()
april_cnt.columns = ['file_id','tm_hour','frequency']
april_cnt = april_cnt.pivot_table(values='frequency',index=['file_id'],columns='tm_hour',aggfunc='sum').reset_index()

del april
gc.collect();

may = pd.read_hdf('may.h5', 'df')
may_cnt = may[['file_id','tm_hour','timestamp']].groupby(['file_id','tm_hour']).count().reset_index()
may_cnt.columns = ['file_id','tm_hour','frequency']
may_cnt = may_cnt.pivot_table(values='frequency',index=['file_id'],columns='tm_hour',aggfunc='sum').reset_index()

del may
gc.collect();
```

```python
data = pd.concat([march_cnt,april_cnt],axis=0)
data = pd.concat([data,may_cnt],axis=0)
```

```python
data = data.groupby('file_id').sum().reset_index()
data.columns = ['file_id'] + ['h'+str(i) for i in data.columns.tolist() if i is not 'file_id']

data.fillna(0,inplace=True)

data['sum'] = data.iloc[:,1:].sum(axis=1)
for i in range(24):
    i=i+1
    data.iloc[:,i] = (data.iloc[:,i]  / data['sum'])*100

data.drop('sum',axis=1,inplace=True)
data.head(3)
```

```python
data.to_csv('hour_cnt.csv',index=False)
```
