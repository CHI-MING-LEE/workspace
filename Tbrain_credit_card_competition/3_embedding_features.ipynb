## Import

```python
import pandas as pd
from keras.layers import Input, Dense
from keras.models import Model
```

## Read

```python
# pre10有轉one-hot試試看~~~~
# Read data
train_raw = pd.read_csv("train_pre_10.csv")
X_test = pd.read_csv("test_pre_10.csv")

# train_raw = pd.read_csv("train_pre_pure.csv")
# X_test = pd.read_csv("test_pre_pure.csv")

test_raw = pd.read_csv("test.csv")

X = train_raw.drop('fraud_ind', axis=1)
y = train_raw['fraud_ind']

print(train_raw.shape)
print(X.shape)
print(X_test.shape)
print(test_raw.shape)

all_features = pd.concat([X, X_test], axis=0)

print(all_features.shape)
```

```python
binary_columns = []
for col in all_features.columns.values.tolist():
    if all_features[col].unique().shape[0] <= 2:
        print(col)
        binary_columns.append(col)
binary_columns
```

```python
all_features[binary_columns].head(3)
```

## 全部變為binary

```python
all_features[binary_columns] = all_features[binary_columns].apply(lambda x: x.replace({x.max(): 1, x.min(): 0}))
all_features[binary_columns].head(200)
```

## One-hot後取DNN latent vector來用

```python
encoding_dim = 20
input_dim = 52

input_features = Input(shape=(input_dim,))

encoded = Dense(64, activation='relu')(input_features)
encoded = Dense(128, activation='relu')(encoded)
# encoded = Dense(256, activation='relu')(encoded)
encoded = Dense(encoding_dim, activation='relu')(encoded)

# decoded = Dense(256, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(64, activation='relu')(decoded)
output = Dense(input_dim, activation='linear')(decoded)

# this model maps an input to its reconstruction
autoencoder = Model(input_features, output)
# this model maps an input to its encoded representation
encoder = Model(input_features, encoded)

autoencoder.summary()
encoder.summary()
```

```python
autoencoder.compile(optimizer='adam', loss='mse')

autoencoder.fit(all_features[binary_columns], all_features[binary_columns],
                epochs=50,
                batch_size=1024,
                shuffle=True)
                # validation_data=(x_test, x_test))
```

```python
one_hot_to_latent_features = encoder.predict(all_features[binary_columns])
one_hot_to_latent_features = pd.DataFrame(one_hot_to_latent_features)
one_hot_to_latent_features.shape
```

```python
one_hot_to_latent_features.apply(lambda x: x.mean())  ## 0 & 8 & 11刪除
del one_hot_to_latent_features[0]
del one_hot_to_latent_features[8]
del one_hot_to_latent_features[11]
```

```python
one_hot_to_latent_features.columns =["latent_" + str(no) for no in range(one_hot_to_latent_features.shape[1])]
one_hot_to_latent_features.head()
```

```python
one_hot_to_latent_features.to_csv("one_hot_to_latent_features.csv", index=None)
```
